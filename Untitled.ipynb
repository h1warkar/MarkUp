```python
from tkinter import *
from PIL import ImageTk, Image
from tkinter import font
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg)
import pandas as pd

sui = Tk()
#sui.geometry('800x600')
sui.attributes('-fullscreen', True)
sui.title('Mark Up')

#File for Spreadsheet
file = pd.read_excel('MarkSheet_Dataset1.xlsx')

#Making the figures for the graph
fig = Figure(figsize=(3.75,3.75), dpi=100)
fig1 = Figure(figsize=(3.75,3.75), dpi=100)


#User defined functions
#start up window


def _sum(arr):

    sum = 0

    for i in arr:
        sum = sum + i
 
    return(sum)

def leave():
     sui.destroy()

def next_screen(event):
    #Background switch
    photolabel1.place_forget()
    photolabel2.place(x=0,y=0)
    
    #Un-place the start window
    student_name_entry.place_forget()
    go_button.place_forget()
    
    #Spreadsheet values
    student = student_name_entry.get()
    student_use = student.lower()
    
    xvalues = file[f'{student_use}'].values
    yvalues = file['Name of Student'].values
    
    xpoint1 = xvalues[0]
    xpoint2 = xvalues[-1]
    xpoint3 = xvalues[-2]
    
    ypoint1 = yvalues[0]
    ypoint2 = yvalues[-1]
    
    xpoints = [xpoint1, xpoint2]
    ypoints = [ypoint1, ypoint2]
    
    classx = file['class average'].values
    
    #Graph 1
    plot1 = fig.add_subplot(111)
    plot1.plot(yvalues, xvalues)
    plot1.plot(yvalues, classx)
    
    plot1.set_xlabel('Grade')
    plot1.set_ylabel('Marks')
    
    plot1.set_title(f"{student}'s marks through the years")
    
    fig.legend([f"{student}",'Class average'], loc='lower right')
    
    canvas = FigureCanvasTkAgg(fig, master=sui)
    canvas.draw()
    canvas.get_tk_widget().place(relx=0.02, rely=0.02)
    
    #Graph 2
    plot2 = fig1.add_subplot(111)
    plot2.plot(yvalues, xvalues)
    plot2.plot(ypoints, xpoints)
    
    plot2.set_xlabel('Grade')
    plot2.set_ylabel('Marks')
    
    plot2.set_title(f"{student}'s improvement graph")
    
    fig1.legend([ f"{student}",'Improvement'], loc='lower right')
    
    canvas1 = FigureCanvasTkAgg(fig1, master=sui)
    canvas1.draw()
    canvas1.get_tk_widget().place(relx=0.02, rely=0.52)
    
    #Improvement frame
    imp = Frame(sui, width=800, height=375,  highlightthickness=2, bg='black')
    imp.place(x=500, y=15)
    
    imp_5_year = int(((xpoint2-xpoint1)/xpoint1)*100)
    xpoint3 = xvalues[-2]
    imp_2_year = int(((xpoint2-xpoint3)/xpoint3)*100)
    total_marks = _sum(xvalues)
    grade_len = len(xvalues)
    average_marks = total_marks/grade_len
    
    
    imp5_title_label = Label(imp, text=f'Improvement of {student} over last 5 years: {imp_5_year}%', bg='black', fg="white", font=(16))
    imp5_title_label.place(x=50, y=40)
    imp2_title_label = Label(imp, text=f'Improvement of {student} over last 2 years: {imp_2_year}%', bg='black', fg="white", font=(16))
    imp2_title_label.place(x=50, y=60)
    average_title_label = Label(imp, text=f'Average marks of {student} over last 5 years: {average_marks}', bg='black', fg="white", font=(16))
    average_title_label.place(x=50, y=80)
    
    #Status frame
    predictor = Frame(sui, width=800, height=340, highlightthickness=2, bg='black')
    predictor.place(x=500, y=400)
    
    pre_5_year = int(xpoint2+(xpoint2-xpoint1))
    if pre_5_year >= 100:
         pre_5_year = 100
    else:
         pass
    pre_2_year = int(xpoint2+(xpoint2-xpoint3))
    if pre_2_year >= 100:
         pre_2_year = 100
    else:
         pass
    p5_imp = int(xpoint2+(5*(xpoint2/100)))
    p10_imp = int(xpoint2+(10*(xpoint2/100)))
    p20_imp = int(xpoint2+(20*(xpoint2/100)))
    
    p5_imp_label = Label(imp, text=f"Marks of {student} if they had a 5% improvement would be: {p5_imp}", bg='black', fg='white', font=(10))
    p5_imp_label.place(x=50, y=100)
    p10_imp_label = Label(imp, text=f"Marks of {student} if they had a 10% improvement would be: {p10_imp}", bg='black', fg='white', font=(10))
    p10_imp_label.place(x=50, y=120)
    p20_imp_label = Label(imp, text=f"Marks of {student} if they had a 20% improvement would be: {p20_imp}", bg='black', fg='white', font=(10))
    p20_imp_label.place(x=50, y=140)
    pre5_label = Label(imp, text=f"Prediction of {student} if they continued in their previous 5 years trend: {pre_5_year}", bg='black', fg='white', font=(10))
    pre5_label.place(x=50, y=160)
    pre2_label = Label(imp, text=f"Prediction of {student} if they continued in their previous 2 years trend: {pre_2_year}", bg='black', fg='white', font=(10))
    pre2_label.place(x=50, y=180)

sui.bind('<Return>', next_screen)
def next_screen1():
    #Background switch
    photolabel1.place_forget()
    photolabel2.place(x=0,y=0)
    
    #Un-place the start window
    student_name_entry.place_forget()
    go_button.place_forget()
    
    #Spreadsheet values
    student = student_name_entry.get()
    student_use = student.lower()
    
    xvalues = file[f'{student_use}'].values
    yvalues = file['Name of Student'].values
    
    xpoint1 = xvalues[0]
    xpoint2 = xvalues[-1]
    xpoint3 = xvalues[-2]
    
    ypoint1 = yvalues[0]
    ypoint2 = yvalues[-1]
    
    xpoints = [xpoint1, xpoint2]
    ypoints = [ypoint1, ypoint2]
    
    classx = file['class average'].values
    
    #Graph 1
    plot1 = fig.add_subplot(111)
    plot1.plot(yvalues, xvalues)
    plot1.plot(yvalues, classx)
    
    plot1.set_xlabel('Grade')
    plot1.set_ylabel('Marks')
    
    plot1.set_title(f"{student}'s marks through the years")
    
    fig.legend([f"{student}",'Class average'], loc='lower right')
    
    canvas = FigureCanvasTkAgg(fig, master=sui)
    canvas.draw()
    canvas.get_tk_widget().place(relx=0.02, rely=0.02)
    
    #Graph 2
    plot2 = fig1.add_subplot(111)
    plot2.plot(yvalues, xvalues)
    plot2.plot(ypoints, xpoints)
    
    plot2.set_xlabel('Grade')
    plot2.set_ylabel('Marks')
    
    plot2.set_title(f"{student}'s improvement graph")
    
    fig1.legend([ f"{student}",'Improvement'], loc='lower right')
    
    canvas1 = FigureCanvasTkAgg(fig1, master=sui)
    canvas1.draw()
    canvas1.get_tk_widget().place(relx=0.02, rely=0.52)
    
    #Improvement frame
    imp = Frame(sui, width=800, height=375,  highlightthickness=2, bg='#007db5')
    imp.place(x=600, y=25)
    
    imp_5_year = int(((xpoint2-xpoint1)/xpoint1)*100)
    xpoint3 = xvalues[-2]
    imp_2_year = int(((xpoint2-xpoint3)/xpoint3)*100)
    total_marks = _sum(xvalues)
    grade_len = len(xvalues)
    average_marks = total_marks/grade_len
    
    
    imp5_title_label = Label(imp, text=f'Improvement of {student} over last 5 years: {imp_5_year}%', bg='#007db5', fg="white", font=(16))
    imp5_title_label.place(x=50, y=40)
    imp2_title_label = Label(imp, text=f'Improvement of {student} over last 2 years: {imp_2_year}%', bg='#007db5', fg="white", font=(16))
    imp2_title_label.place(x=50, y=60)
    average_title_label = Label(imp, text=f'Average marks of {student} over last 5 years: {average_marks}', bg='#007db5', fg="white", font=(16))
    average_title_label.place(x=50, y=80)
    
    #Status frame
    status = Frame(sui, width=800, height=340, highlightthickness=2, bg='black')
    status.place(x=600, y=540)
    
    pre_5_year = int(xpoint2+(xpoint2-xpoint1))
    if pre_5_year >= 100:
         pre_5_year = 100
    else:
         pass
    pre_2_year = int(xpoint2+(xpoint2-xpoint3))
    if pre_2_year >= 100:
         pre_2_year = 100
    else:
         pass
    p5_imp = int(xpoint2+(5*(xpoint2/100)))
    p10_imp = int(xpoint2+(10*(xpoint2/100)))
    p20_imp = int(xpoint2+(20*(xpoint2/100)))
    
    p5_imp_label = Label(imp, text=f"Marks of {student} if they had a 5% improvement would be: {p5_imp}", bg='#007db5', fg='white', font=(10))
    p5_imp_label.place(x=50, y=100)
    p10_imp_label = Label(imp, text=f"Marks of {student} if they had a 10% improvement would be: {p10_imp}", bg='#007db5', fg='white', font=(10))
    p10_imp_label.place(x=50, y=120)
    p20_imp_label = Label(imp, text=f"Marks of {student} if they had a 20% improvement would be: {p20_imp}", bg='#007db5', fg='white', font=(10))
    p20_imp_label.place(x=50, y=140)
    pre5_label = Label(imp, text=f"Prediction of {student} if they continued in their previous 5 years trend: {pre_5_year}", bg='#007db5', fg='white', font=(10))
    pre5_label.place(x=50, y=160)
    pre2_label = Label(imp, text=f"Prediction of {student} if they continued in their previous 2 years trend: {pre_2_year}", bg='#007db5', fg='white', font=(10))
    pre2_label.place(x=50, y=180)

#Background
photo1=PhotoImage(file='background{M}.png', master=sui)
photolabel1= Label(sui, image=photo1)
photolabel1.pack()

photo2=PhotoImage(file='background{E}.png', master=sui)
photolabel2= Label(sui, image=photo2)
photolabel2.pack()
student_name_entry = Entry(sui, width=38, font=("Times New Roman", 16))
student_name_entry.place(x=480, y=280)


go_button = Button(sui, text='GO!!!', command=next_screen1, width=12)
go_button.place(x=640, y=350)

#Copyright: AH and DI


#Exit button
exit_button = Button(sui, text='Exit', command=leave, width=12)
exit_button.place(x=650 , y=740)

sui.mainloop()

```
